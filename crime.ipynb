Расчитываем матрицу корреляции

```python
corr_mat = data.corr()
threshold = 0.9
for i, name1 in enumerate(data.columns):
    for name2 in data.columns[i+1:]:
        if math.fabs(corr_mat[name1][name2]) > threshold:
            print name1, '-', name2, ':', corr_mat[name1][name2]
```

Определяем сильно коррелирующие значения
Можно заметить, что с течением времени количетсво населения в стране растет. Растет также и общее количетсво преступлений

```python
corr_mat = data.corr()
threshold = 0.2
for i, name1 in enumerate(data.columns):
    for name2 in data.columns[i+1:]:
        if math.fabs(corr_mat[name1][name2]) < threshold:
            print name1, '-', name2, ':', corr_mat[name1][name2]
```

```python
data = data.drop('Year', axis = 1)
```

```python
X = data.drop(('murder'), axis = 1) # выбрасываем столбец 'class'
X = (X - X.mean(axis = 0))/X.std(axis = 0)
```

```python
murder_count = data['murder']
murder_count.describe()
```

```python
murder_count.unique()
```

4 различных значения

```python
print '= 1: ', np.count_nonzero(murder_count == 1)
print '> 1: ', np.count_nonzero(murder_count > 1)
```

```python
y = murder_count == 1
y.unique()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
print y_train
```

#kNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
#knn.set_params(n_neighbors=10)
knn.fit(X_train, y_train)
```

```python
knn
```

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

```python
err_test = 1 - knn.score(X_test, y_test) # Другой способ
```

```python
err_test
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```
